```python
# === TEST ISTOTNOŚCI STATYSTYCZNEJ: 5x2cv F-Alpaydina ===
# Główna procedura testowa zgodnie z Alpaydin (1999), oparta na 5x2cv (Dietterich 1998).
# Wyniki z 5x2cv są wczytywane z osobnego notebooka train_5x2cv.ipynb.
#
# Test wykorzystuje 5 powtórzeń × 2 podziały (50/50) = 10 wartości metryki na architekturę,
# z odpowiednią strukturą wariancji wewnątrz powtórzeń.
# Rozkład: F(10, 5).

from scipy.stats import f as f_dist
from glob import glob

RESULTS_5X2_DIR = f'{DRIVE_BASE}/results/5x2cv'

def load_5x2cv_metrics(metric_key):
    """Wczytuje wartości metryki ze wszystkich powtorzeń 5x2cv.

    Zwraca:
        p1_a, p2_a: shape (5,) - wartości dla ResNet50 w split_a i split_b (po jednym z każdego powt.)
        p1_b, p2_b: shape (5,) - wartości dla VGG16 w split_a i split_b
    """
    p1_resnet, p2_resnet = [], []
    p1_vgg, p2_vgg = [], []

    for rep_idx in range(5):
        rep_dir = f'{RESULTS_5X2_DIR}/rep_{rep_idx}'
        if not os.path.exists(rep_dir):
            return None

        try:
            with open(f'{rep_dir}/split_a/resnet50_metrics.json') as f:
                p1_resnet.append(json.load(f)[metric_key])
            with open(f'{rep_dir}/split_b/resnet50_metrics.json') as f:
                p2_resnet.append(json.load(f)[metric_key])
            with open(f'{rep_dir}/split_a/vgg16_metrics.json') as f:
                p1_vgg.append(json.load(f)[metric_key])
            with open(f'{rep_dir}/split_b/vgg16_metrics.json') as f:
                p2_vgg.append(json.load(f)[metric_key])
        except FileNotFoundError:
            return None

    return (np.array(p1_resnet), np.array(p2_resnet),
            np.array(p1_vgg),    np.array(p2_vgg))


def alpaydin_5x2cv_f_test(p1_a, p2_a, p1_b, p2_b):
    """Test F-Alpaydina (1999) dla porównania dwóch algorytmów uczących.

    Argumenty:
        p1_a, p2_a: wartości metryki dla algorytmu A w split 1 i 2 każdego z 5 powtórzeń, kształt (5,)
        p1_b, p2_b: wartości metryki dla algorytmu B w split 1 i 2 każdego z 5 powtórzeń, kształt (5,)

    Zwraca:
        F: statystyka F
        p_value: p-wartość z rozkładu F(10, 5)
    """
    p1 = p1_a - p1_b  # różnice w split 1 dla każdego powtórzenia
    p2 = p2_a - p2_b  # różnice w split 2 dla każdego powtórzenia
    p_mean = (p1 + p2) / 2.0
    s_sq = (p1 - p_mean)**2 + (p2 - p_mean)**2  # wariancja w powtórzeniu

    numerator = np.sum(p1**2 + p2**2)
    denominator = 2.0 * np.sum(s_sq)

    if denominator < 1e-12:
        return np.inf, 0.0

    F = numerator / denominator
    p_value = 1.0 - f_dist.cdf(F, 10, 5)
    return F, p_value


# === URUCHOMIENIE TESTU ===
print('=' * 70)
print('TEST F-ALPAYDINA (5x2cv) - GŁÓWNA PROCEDURA TESTOWA')
print('=' * 70)

has_5x2cv = all(
    os.path.exists(f'{RESULTS_5X2_DIR}/rep_{i}') for i in range(5)
)

if not has_5x2cv:
    print('UWAGA: brak wyników 5x2cv. Uruchom najpierw train_5x2cv.ipynb dla wszystkich')
    print('powtórzeń (REP_IDX = 0..4). Test F-Alpaydina pominięty.')
else:
    f_test_results = {}
    for metric, label in metric_labels.items():
        data = load_5x2cv_metrics(metric)
        if data is None:
            print(f'{label:>25}: brak kompletnych danych 5x2cv')
            continue

        p1_r, p2_r, p1_v, p2_v = data
        F, p_value = alpaydin_5x2cv_f_test(p1_r, p2_r, p1_v, p2_v)

        mean_diff_pp = ((p1_v.mean() + p2_v.mean())/2 - (p1_r.mean() + p2_r.mean())/2) * 100
        direction = 'VGG16 > ResNet50' if mean_diff_pp > 0 else 'ResNet50 > VGG16'

        if p_value < 0.05:
            verdict = f'ISTOTNE STATYSTYCZNIE (p={p_value:.4f}, α=0,05)'
        else:
            verdict = f'brak istotności statystycznej (p={p_value:.4f})'

        print(f'{label:>25}: różnica {mean_diff_pp:+.2f}pp ({direction})')
        print(f'{"":>25}  F(10, 5) = {F:.3f}')
        print(f'{"":>25}  {verdict}')

        f_test_results[metric] = {
            'F': float(F),
            'p_value': float(p_value),
            'mean_diff_pp': float(mean_diff_pp),
            'direction': direction,
        }

    with open(f'{OUTPUT_DIR}/f_test_alpaydin.json', 'w') as f:
        json.dump(f_test_results, f, indent=2)


# === BOOTSTRAP CI DLA RÓŻNICY METRYK (z 5-fold CV - bez nowych trenowań) ===
print('\n' + '=' * 70)
print('BOOTSTRAP 95% CI dla różnicy metryk (VGG16 - ResNet50)')
print('z istniejących wyników 5-fold CV (10000 resamples)')
print('=' * 70)

N_BOOTSTRAP = 10000
RNG = np.random.default_rng(42)

def bootstrap_diff_ci(values_a, values_b, n_boot=N_BOOTSTRAP, ci=0.95):
    """Bootstrap percentile CI dla różnicy średnich z par 5-fold CV.

    Bierze 5 par (a_i, b_i) z foldów, resampluje z zwracaniem,
    oblicza średnią różnicę dla każdego resampla.
    """
    n_folds = len(values_a)
    diffs = values_b - values_a  # różnica VGG16 - ResNet50 w każdym foldzie

    boot_diffs = np.empty(n_boot)
    for i in range(n_boot):
        idx = RNG.integers(0, n_folds, size=n_folds)
        boot_diffs[i] = diffs[idx].mean()

    alpha = (1 - ci) / 2
    lo = np.percentile(boot_diffs, alpha * 100)
    hi = np.percentile(boot_diffs, (1 - alpha) * 100)
    return diffs.mean(), lo, hi


bootstrap_results = {}
for metric, label in metric_labels.items():
    resnet_vals = np.array([r[metric] for r in all_results['ResNet50']])
    vgg_vals = np.array([r[metric] for r in all_results['VGG16']])

    mean_diff, lo, hi = bootstrap_diff_ci(resnet_vals, vgg_vals)
    excludes_zero = (lo > 0) or (hi < 0)
    flag = ' [CI nie zawiera 0]' if excludes_zero else ''

    print(f'{label:>25}: Δ = {mean_diff*100:+.3f}pp  '
          f'95% CI [{lo*100:+.3f}, {hi*100:+.3f}]pp{flag}')

    bootstrap_results[metric] = {
        'mean_diff_pp': float(mean_diff * 100),
        'ci_lo_pp': float(lo * 100),
        'ci_hi_pp': float(hi * 100),
        'excludes_zero': bool(excludes_zero),
    }

with open(f'{OUTPUT_DIR}/bootstrap_ci.json', 'w') as f:
    json.dump(bootstrap_results, f, indent=2)


# === MATCHED-PAIRS RANK-BISERIAL CORRELATION (effect size dla 5-fold CV) ===
print('\n' + '=' * 70)
print('Matched-pairs rank-biserial correlation (effect size, 5-fold CV)')
print('=' * 70)

for metric, label in metric_labels.items():
    resnet_vals = np.array([r[metric] for r in all_results['ResNet50']])
    vgg_vals = np.array([r[metric] for r in all_results['VGG16']])
    diffs = vgg_vals - resnet_vals
    n_pos = (diffs > 0).sum()
    n_neg = (diffs < 0).sum()
    n_total = n_pos + n_neg
    if n_total == 0:
        r_rb = 0.0
    else:
        r_rb = (n_pos - n_neg) / n_total
    print(f'{label:>25}: r_rb = {r_rb:+.3f}  ({n_pos}/{len(diffs)} podziałów VGG16 > ResNet50)')


# === ANALIZA POMOCNICZA: test Wilcoxona (zachowana dla kontekstu historycznego) ===
print('\n' + '=' * 70)
print('ANALIZA POMOCNICZA: test Wilcoxona na 5-fold CV')
print('(uwaga: przy n=5 minimum p-value testu dwustronnego wynosi 0,0625)')
print('=' * 70)

for metric, label in metric_labels.items():
    resnet_vals = np.array([r[metric] for r in all_results['ResNet50']])
    vgg_vals = np.array([r[metric] for r in all_results['VGG16']])

    if np.allclose(resnet_vals, vgg_vals):
        print(f'{label:>25}: wartości identyczne, test nieaplikowalny')
        continue

    stat, p = wilcoxon(resnet_vals, vgg_vals)
    print(f'{label:>25}: W={stat:.1f}, p={p:.4f}')
```

```python
# === GRAD-CAM ===
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_resnet
from tensorflow.keras.applications.vgg16 import preprocess_input as preprocess_vgg
import matplotlib.cm as cm

gradcam_fold_dir = fold_dirs[FOLD_FOR_GRADCAM]
resnet_weights = f'{gradcam_fold_dir}/resnet50_best.h5'
vgg_weights = f'{gradcam_fold_dir}/vgg16_best.h5'

if not (os.path.exists(resnet_weights) and os.path.exists(vgg_weights)):
    print(f'Brak wag dla podziału {FOLD_FOR_GRADCAM+1}.')
    print('Upewnij się że w treningowym notebooku ustawiono SAVE_WEIGHTS=True')
    model_resnet = None
    model_vgg = None
else:
    print(f'Wczytuję wagi z podziału {FOLD_FOR_GRADCAM+1}')
    model_resnet = load_model(resnet_weights)
    model_vgg = load_model(vgg_weights)
```

```python
# === GENEROWANIE MAP GRAD-CAM (POPRAWKA: gradient na logicie) ===
# Zgodnie z oryginalną propozycją Selvaraju et al. (2017), gradient liczony
# jest na surowym wyjściu warstwy klasyfikacyjnej (logicie), a nie na
# wyjściu po funkcji softmax. Liczenie gradientu na softmax prowadzi do
# zjawiska saturacji dla predykcji o wysokiej pewności i może skutkować
# brakiem czytelnych map aktywacji nawet dla obrazów poprawnie sklasyfikowanych.

def make_gradcam(img_array, seq_model, conv_layer_name):
    """Generuje mapę Grad-CAM dla obrazu i modelu.

    Tymczasowo wyłącza aktywację softmax w ostatniej warstwie, żeby gradient
    był liczony na logitach (zgodnie z Selvaraju et al. 2017).
    """
    # Zapamiętaj oryginalną aktywację i wyłącz softmax (linear = identity)
    last_layer = seq_model.layers[-1]
    original_activation = last_layer.activation
    last_layer.activation = tf.keras.activations.linear

    try:
        backbone = seq_model.layers[0]
        inputs = backbone.input
        conv_out = backbone.get_layer(conv_layer_name).output
        x = backbone.output
        for layer in seq_model.layers[1:]:
            x = layer(x)
        # x to teraz LOGITY (bo wyłączyliśmy softmax w ostatniej warstwie)
        grad_model = tf.keras.Model(inputs=inputs, outputs=[conv_out, x])

        with tf.GradientTape() as tape:
            conv_outputs, logits = grad_model(img_array)
            pred_idx = tf.argmax(logits[0])
            loss = logits[:, pred_idx]

        grads = tape.gradient(loss, conv_outputs)
        pooled = tf.reduce_mean(grads, axis=(0, 1, 2))
        heatmap = tf.reduce_mean(tf.multiply(pooled, conv_outputs[0]), axis=-1)
        heatmap = tf.maximum(heatmap, 0)
        heatmap = heatmap / (tf.reduce_max(heatmap) + 1e-8)
    finally:
        # Przywróć oryginalną aktywację softmax
        last_layer.activation = original_activation

    return heatmap.numpy()


sample_classes = ['Tomato_Bacterial_spot', 'Tomato_Late_blight',
                   'Tomato_healthy', 'Potato___Early_blight']

if model_resnet is not None and model_vgg is not None:
    jet = plt.colormaps['jet']
    fig, axes = plt.subplots(len(sample_classes), 3, figsize=(15, 5 * len(sample_classes)))

    for idx, cls in enumerate(sample_classes):
        cls_path = os.path.join(SRC, cls)
        if not os.path.exists(cls_path):
            print(f'Pomijam klasę {cls}: brak ścieżki')
            continue
        img_path = os.path.join(cls_path, sorted(os.listdir(cls_path))[0])
        orig = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
        orig_arr = tf.keras.preprocessing.image.img_to_array(orig)

        axes[idx, 0].imshow(orig)
        axes[idx, 0].set_title(f'Oryginał\n{cls}', fontsize=9)
        axes[idx, 0].axis('off')

        for col, (mdl, preproc, layer_name, mname) in enumerate([
            (model_resnet, preprocess_resnet, 'conv5_block3_out', 'ResNet50'),
            (model_vgg, preprocess_vgg, 'block5_conv3', 'VGG16'),
        ], start=1):
            img_in = preproc(np.expand_dims(orig_arr.copy(), axis=0))
            heatmap = make_gradcam(img_in, mdl, layer_name)
            heatmap = tf.image.resize(heatmap[..., None], (224, 224)).numpy()[..., 0]
            jet_colors = jet(heatmap)[..., :3]
            overlay = 0.5 * (orig_arr / 255.0) + 0.5 * jet_colors

            axes[idx, col].imshow(overlay)
            axes[idx, col].set_title(f'{mname} Grad-CAM\n(podział {FOLD_FOR_GRADCAM+1})', fontsize=9)
            axes[idx, col].axis('off')

    plt.tight_layout()
    plt.savefig(f'{OUTPUT_DIR}/fig5_gradcam.png', dpi=200, bbox_inches='tight')
    plt.show()
else:
    print('Pominięto Grad-CAM, brak modeli.')
```

```python
# === ROZSZERZONY PRZEGLĄD GRAD-CAM: 2 OBRAZY × 15 KLAS ==========================

import os
import csv
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold

RANDOM_SEED = 42
N_FOLDS = 5

if model_resnet is None or model_vgg is None:
    raise RuntimeError('Brak modeli — uruchom najpierw komórki Grad-CAM powyżej.')

# === ODTWORZENIE PODZIAŁU TESTOWEGO DLA WYBRANEGO FOLDA =================
rows = []
for cls in sorted(os.listdir(SRC)):
    cls_path = os.path.join(SRC, cls)
    if not os.path.isdir(cls_path):
        continue
    for f in sorted(os.listdir(cls_path)):
        rows.append({'filepath': os.path.join(cls_path, f), 'class': cls})
df = pd.DataFrame(rows)

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_SEED)
fold_splits = list(skf.split(df, df['class']))
_, test_idx = fold_splits[FOLD_FOR_GRADCAM]
df_test = df.iloc[test_idx].reset_index(drop=True)

print(f'Zbiór testowy podziału {FOLD_FOR_GRADCAM+1}: {len(df_test)} obrazów')

# === WYBÓR 2 OBRAZÓW NA KLASĘ ========
classes = sorted(df_test['class'].unique())
print(f'Klas w zbiorze testowym: {len(classes)}')

selected = []
for cls in classes:
    cls_files = sorted(df_test[df_test['class'] == cls]['filepath'].tolist())
    if len(cls_files) < 2:
        print(f'  UWAGA: klasa {cls} ma tylko {len(cls_files)} obrazów testowych')
    selected.append({'cls': cls, 'paths': cls_files[:2]})

# === KONTAKTOWE ARKUSZE PER KLASA + KOLEKCJONOWANIE WIERSZY CSV =========
review_dir = f'{OUTPUT_DIR}/gradcam_review'
os.makedirs(review_dir, exist_ok=True)

jet = plt.colormaps['jet']
csv_rows = []

for cls_idx, item in enumerate(selected, start=1):
    cls = item['cls']
    paths = item['paths']
    n_rows = len(paths)
    if n_rows == 0:
        continue

    fig, axes = plt.subplots(n_rows, 3, figsize=(15, 5 * n_rows))
    if n_rows == 1:
        axes = axes[np.newaxis, :]

    for r, img_path in enumerate(paths):
        orig = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
        orig_arr = tf.keras.preprocessing.image.img_to_array(orig)

        axes[r, 0].imshow(orig)
        axes[r, 0].set_title(f'Oryginał\n{os.path.basename(img_path)}', fontsize=9)
        axes[r, 0].axis('off')

        for col, (mdl, preproc, layer_name, mname) in enumerate([
            (model_resnet, preprocess_resnet, 'conv5_block3_out', 'ResNet50'),
            (model_vgg,    preprocess_vgg,    'block5_conv3',     'VGG16'),
        ], start=1):
            img_in = preproc(np.expand_dims(orig_arr.copy(), axis=0))
            heatmap = make_gradcam(img_in, mdl, layer_name)
            heatmap = tf.image.resize(heatmap[..., None], (224, 224)).numpy()[..., 0]
            jet_colors = jet(heatmap)[..., :3]
            overlay = 0.5 * (orig_arr / 255.0) + 0.5 * jet_colors

            axes[r, col].imshow(overlay)
            axes[r, col].set_title(f'{mname} Grad-CAM\n(podział {FOLD_FOR_GRADCAM+1})', fontsize=9)
            axes[r, col].axis('off')

        csv_rows.append({
            'klasa': cls,
            'obraz': os.path.basename(img_path),
            'kategoria_resnet50': '',  # do wypełnienia: A / B / C / D
            'kategoria_vgg16':    '',  # do wypełnienia: A / B / C / D
            'uwagi':              '',  # opcjonalne notatki
        })

    plt.suptitle(f'Klasa: {cls}', fontsize=14, y=1.00)
    plt.tight_layout()
    out_path = f'{review_dir}/{cls_idx:02d}_{cls}.png'
    plt.savefig(out_path, dpi=160, bbox_inches='tight')
    plt.close(fig)
    print(f'  zapisano: {out_path}')

# === TEMPLATE CSV DO RĘCZNEGO KATALOGOWANIA ============================
csv_path = f'{review_dir}/_katalogowanie.csv'
with open(csv_path, 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(
        f, fieldnames=['klasa', 'obraz', 'kategoria_resnet50', 'kategoria_vgg16', 'uwagi']
    )
    writer.writeheader()
    for row in csv_rows:
        writer.writerow(row)

print(f'\nTemplate katalogowania: {csv_path}')
print(f'Map do oceny: {len(csv_rows)} obrazów × 2 modele = {len(csv_rows)*2} map.')
print()
print('Kategorie:')
print('  A — aktywacja skupiona na zmianie chorobowej')
print('      (dla klas zdrowych: aktywacja na obszarze liścia)')
print('  B — aktywacja rozproszona po liściu (bez wyraźnego skupienia)')
print('  C — aktywacja głównie na tle / poza obszarem liścia')
print('  D — brak czytelnej aktywacji (mapa głównie ciemnoniebieska)')
```
